### First, let's build an xgboost classifier model with the data as is (without undersamping or oversampling).
#### Create a train-test split.

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns=['Class'])
y = df['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)
```

We will standard-scale the `Time` and  `Amount` features.

```python
from sklearn.preprocessing import StandardScaler
sclr = StandardScaler()

scaled = pd.DataFrame(sclr.fit_transform(X_train[['Time', 'Amount']]), index=X_train.index)

X_train['Time'] = scaled[0]
X_train['Amount'] = scaled[1]

test_scaled = pd.DataFrame(sclr.transform(X_test[['Time', 'Amount']]), index=X_test.index)

X_test['Time'] = test_scaled[0]
X_test['Amount'] = test_scaled[1]
```

### Let's try to handle outliers.

We will treat data lying outside the respective interquartile ranges in each column to be outliers.

```python
Q1 = X_train.quantile(0.25)
Q3 = X_train.quantile(0.75)

IQR = Q3 - Q1

nad = X_train[(X_train > Q1 - 1.5 * IQR) & (X_train < Q3 + 1.5 * IQR)]

(nad.shape[0] - nad.dropna().shape[0])/nad.shape[0]
```

Almost 50% of the rows contain outliers. So we won't drop it. We will treat them as missing values and fill them with their column means.

```python
prep_X_train = nad.fillna(nad.median())
```

Do similar with the test data.

```python
Q1 = X_test.quantile(0.25)
Q3 = X_test.quantile(0.75)

IQR = Q3 - Q1

nad = X_test[(X_test > Q1 - 1.5 * IQR) & (X_test < Q3 + 1.5 * IQR)]

prep_X_test = nad.fillna(nad.median())
```

Build an xgboost classifier model with default parameters.

```python
from xgboost import XGBClassifier

xgb = XGBClassifier()

xgb.fit(prep_X_train, y_train)

y_hat = xgb.predict(prep_X_test)
```

Let's evaluate our model.

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, recall_score, precision_score, roc_auc_score

cm = confusion_matrix(y_test, y_hat)

ConfusionMatrixDisplay(confusion_matrix = cm).plot()
```

```python
print(f'Test Accuracy: {xgb.score(prep_X_test, y_test)}\
      \nTest Precision: {precision_score(y_test, y_hat)}\
      \nTest Recall: {recall_score(y_test, y_hat)}\
      \nTest ROC AUC Score: {roc_auc_score(y_test, y_hat)}')
```

These numbers are very bad. Let's run the same model without preprocessing.

```python
# We will save the previous model before proceeding.

with open(r"..\models\xgb-preprocess-all.pkl", "wb") as file:
    pk.dump(xgb, file)
```

So preprocessing/outlier-treatment is not required. Infact, it makes our model worse.

## We will now try undersampling.

```python
y.value_counts()
```

We will take a sample size equal to occurence of class `1` from the `0` class. And then concatenate it with the entirety of the `1` class.

```python
sample0 = df[df['Class'] == 0].sample(y.value_counts()[1])

under_df = pd.concat([sample0, df[df['Class'] == 1]]).reset_index(drop=True)

under_df['Class'].value_counts()
```

Next, we will create the train-test split and proceed as usual.

```python
under_X = under_df.drop(columns=['Class'])
under_y = under_df['Class']

under_X_train, under_X_test, under_y_train, under_y_test = train_test_split(under_X, under_y, test_size=0.2, random_state=123)

# We will standard-scale the Time and Amount features.
under_sclr = StandardScaler()

under_scaled = pd.DataFrame(under_sclr.fit_transform(under_X_train[['Time', 'Amount']]), index=under_X_train.index)

under_X_train['Time'] = under_scaled[0]
under_X_train['Amount'] = under_scaled[1]

under_test_scaled = pd.DataFrame(under_sclr.transform(under_X_test[['Time', 'Amount']]), index=under_X_test.index)

under_X_test['Time'] = under_test_scaled[0]
under_X_test['Amount'] = under_test_scaled[1]


# We will build an xgb classifier
under_xgb = XGBClassifier()

under_xgb.fit(under_X_train, under_y_train)

under_y_hat = under_xgb.predict(under_X_test)

under_cm = confusion_matrix(under_y_test, under_y_hat)

ConfusionMatrixDisplay(confusion_matrix = under_cm).plot()

print(f'Test Accuracy: {under_xgb.score(under_X_test, under_y_test)}\
      \nTest Precision: {precision_score(under_y_test, under_y_hat)}\
      \nTest Recall: {recall_score(under_y_test, under_y_hat)}\
      \nTest ROC AUC Score: {roc_auc_score(under_y_test, under_y_hat)}')
```

This is by far the best model. All scores are above 0.9!!

```python
# We will save the previous model before proceeding.

with open(r"..\models\xgb-under.pkl", "wb") as file:
    pk.dump(under_xgb, file)
```

### Let's also a build a random forest model using hyperparameter tuning.

### We will now try oversampling.

One needs to be careful here. We cannot simply make duplicates of the rows with class `1` to match the occurences of class `0` and concatenate them. This will cause data leakage resulting in high test scores.

So we will do train-test split first and do oversampling on the train data only. This will prevent data leakage.

Also, remember to do standard-scaling before oversampling.

```python
df_train, df_test = train_test_split(df, test_size=0.3, random_state=123)


# We will standard-scale the Time and Amount features.
over_sclr = StandardScaler()

over_scaled = pd.DataFrame(over_sclr.fit_transform(df_train[['Time', 'Amount']]), index=df_train.index)

df_train['Time'] = over_scaled[0]
df_train['Amount'] = over_scaled[1]

over_test_scaled = pd.DataFrame(over_sclr.transform(df_test[['Time', 'Amount']]), index=df_test.index)

df_test['Time'] = over_test_scaled[0]
df_test['Amount'] = over_test_scaled[1]


diff = df_train['Class'].value_counts()[0] - df_train['Class'].value_counts()[1]

sample1 = df_train[df_train['Class'] == 1].sample(diff, replace=True)
over_df_train = pd.concat([df_train, sample1]).reset_index(drop=True)

over_X_train = over_df_train.drop(columns=['Class'])
over_y_train = over_df_train['Class']

over_X_test = df_test.drop(columns=['Class'])
over_y_test = df_test['Class']


# We will build an xgb classifier
over_xgb = XGBClassifier()

over_xgb.fit(over_X_train, over_y_train)

over_y_hat = over_xgb.predict(over_X_test)

over_cm = confusion_matrix(over_y_test, over_y_hat)

ConfusionMatrixDisplay(confusion_matrix = over_cm).plot()

print(f'Test Accuracy: {over_xgb.score(over_X_test, over_y_test)}\
      \nTest Precision: {precision_score(over_y_test, over_y_hat)}\
      \nTest Recall: {recall_score(over_y_test, over_y_hat)}\
      \nTest ROC AUC Score: {roc_auc_score(over_y_test, over_y_hat)}')
```

This is showing a little lower performance than the corresponding unbalanced model.

```python
# We will save the previous model before proceeding.

with open(r"..\models\xgb-over.pkl", "wb") as file:
    pk.dump(over_xgb, file)
```

### Let's also a build a random forest model using hyperparameter tuning.

```python
model = RandomForestClassifier()

parameters = {'n_estimators': [10, 50], 'max_depth': [2, 5, 10]}

grid_search = GridSearchCV(model, parameters, cv = 5, scoring = "roc_auc", verbose=3.5)
grid_search.fit(over_X_train, over_y_train)

over_rf = grid_search.best_estimator_
over_rf.fit(over_X_train, over_y_train)

over_y_hat = over_rf.predict(over_X_test)

over_cm = confusion_matrix(over_y_test, over_y_hat)
ConfusionMatrixDisplay(confusion_matrix = over_cm).plot()

print(f'Test Accuracy: {over_rf.score(over_X_test, over_y_test)}\
      \nTest Precision: {precision_score(over_y_test, over_y_hat)}\
      \nTest Recall: {recall_score(over_y_test, over_y_hat)}\
      \nTest ROC AUC Score: {roc_auc_score(over_y_test, over_y_hat)}')
```

We see that the model with oversampled data doesn't do better than the undersampled data. This makes sense because our data is highly imbalanced.

```python
# We will save the previous model.

with open(r"..\models\rf_gridcv-over.pkl", "wb") as file:
    pk.dump(over_rf, file)
```
